### Preprocessing

```python
from sklearn.model_selection import train_test_split
```

```python
ohe=pd.get_dummies(data=df, columns=['Status'])
```

```python
ohe.head()
```

```python
df_new = ohe.drop(columns=['Country', 'Year'])
```

```python
# Separating the target column from the data
X = df_new.drop('Life expectancy ', axis=1)
y = df_new['Life expectancy ']
```

```python
X.columns
```

```python
y.shape
```

```python
# Doing train test split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=4)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Creating holders to store the model performance results
ML_Model = []
R2_train = []
R2_test = []
MSE_train = []
MSE_test = []

#function to call for storing the results
def storeResults(model, a,b,c,d):
  ML_Model.append(model)
  R2_train.append(round(a, 3))
  R2_test.append(round(b, 3))
  MSE_train.append(round(c,3)) 
  MSE_test.append(round(d,3))
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
# Removing insignificant feature
X_train.drop(columns=['Population'], inplace=True)
```

```python
X_test.drop(columns = ['Population'], inplace=True)
```

```python
X_train.columns
```

```python
y_train.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_train_lr = lr.predict(X_train)
y_test_lr = lr.predict(X_test)
```

```python
R2_train_lr = r2_score(y_train, y_train_lr)
R2_test_lr = r2_score(y_test, y_test_lr)
MSE_train_lr = mean_squared_error(y_train, y_train_lr)
MSE_test_lr = mean_squared_error(y_test, y_test_lr)
```

```python
print("Linear regression: R2 score on training Data: {:.3f}".format(R2_train_lr))
print("Linear regression: R2 score on test Data: {:.3f}".format(R2_test_lr))
print("Linear regression: MSE on training Data: {:.3f}".format(MSE_train_lr))
print("Linear regression: MSE on test Data: {:.3f}".format(MSE_test_lr))
```

```python
# Storing results
storeResults('Linear regression', R2_train_lr, R2_test_lr,MSE_train_lr,MSE_test_lr)
```

```python
print("Polynomial regression: R2 score on training Data: {:.3f}".format(R2_train_poly))
print("Polynomial regression: R2 score on test Data: {:.3f}".format(R2_test_poly))
print("Polynomial regression: MSE on training Data: {:.3f}".format(MSE_train_poly))
print("Polynomial regression: MSE on test Data: {:.3f}".format(MSE_test_poly))
```

```python
storeResults('Polynomial regression', R2_train_poly, R2_test_poly, MSE_train_poly, MSE_test_poly)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [2,4,8,10],
    'max_features': [0.2,0.4,0.75,1.0],
    'min_samples_split': [2,5,10,15,20,25],
    'min_samples_leaf': [2,6,10,15],
    'criterion': ['squared_error', 'absolute_error']
}
```

```python
search = GridSearchCV(estimator=DecisionTreeRegressor(), param_grid=param_grid, cv=10, n_jobs=-1, verbose=1, scoring='r2')
search.fit(X_train, y_train)

print('Best features:', search.best_params_)
```

```python
print("Decision Tree regression: R2 score on training Data: {:.3f}".format(R2_train_dt))
print("Decision Tree regression: R2 score on test Data: {:.3f}".format(R2_test_dt))
print("Decision Tree regression: MSE on training Data: {:.3f}".format(MSE_train_dt))
print("Decision Tree regression: MSE on test Data: {:.3f}".format(MSE_test_dt))
```

```python
# Store results
storeResults('Decision tree regressor', R2_train_dt, R2_test_dt, MSE_train_dt, MSE_test_dt)
```

### SVR

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR(kernel='rbf'))
])

param_grid = {

    'svr__degree': [2,3],
    'svr__gamma': ['auto', 'scale'],
    'svr__epsilon': [0.1,0.5,1.0],
    'svr__C': [1000,10000]
}

search1 = GridSearchCV(estimator=pipe, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1, scoring='r2')

search1.fit(X_train, y_train)
```

```python
print('Best features:', search1.best_params_)
```

```python
svr = search1.best_estimator_
```

```python
# Finding predicted values
y_train_svr = svr.predict(X_train)
y_test_svr = svr.predict(X_test)
```

```python
R2_train_svr = r2_score(y_train, y_train_svr)
R2_test_svr = r2_score(y_test, y_test_svr)
MSE_train_svr = mean_squared_error(y_train, y_train_svr)
MSE_test_svr = mean_squared_error(y_test, y_test_svr)
```

```python
print("SVR regression: R2 score on training Data: {:.3f}".format(R2_train_svr))
print("SVR regression: R2 score on test Data: {:.3f}".format(R2_test_svr))
print("SVR regression: MSE on training Data: {:.3f}".format(MSE_train_svr))
print("SVR regression: MSE on test Data: {:.3f}".format(MSE_test_svr))
```

```python
storeResults('SVR', R2_train_svr, R2_test_svr, MSE_train_svr, MSE_test_svr)
```

```python
cv_scores = cross_val_score(svr, X, y, cv=10, scoring='r2')
print(f"Mean R2: {cv_scores.mean():.4f}")
print(f"Std Dev: {cv_scores.std():.4f}")
```
